```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing(as_frame= True).frame
```

```python
# which means of getting 5 first rows we can set amount of rows too in the .head()
data.head(10)

# which is for getting 5 last rows  and we can set amount of rows in the .tail()
data.tail()
```

```python
# .sample for picking up an random row  also we can set an amount .sample()
data.sample()
```

```python
# data.columns   # gives u the names of columns

data = list(data.columns) # it will just give u the columns name in the list format
```

```python
# display limit
pd.options.display.max_columns =5

data 
```

```python
# for getting the "info()" in a concise manner

data.info()
```

```python
# Summarize  format
data.describe()
```

```python
# to get Particular column information u can do like this
data.HouseAge

# or 
data['HouseAge']

# this will give u a series of that column 
```

```python
# now by doiing this we can run statical function or can say do analytical stuff

#mean
data.HouseAge.mean()

#minimum 
data.HouseAge.min()

#maximum
data.HouseAge.max()

#mode
data.HouseAge.mode()

#median
data.HouseAge.median()

#standard deviation
data.HouseAge.std()

#sum 
data.HouseAge.sum()

#count
data.HouseAge.count()

```

```python
#Visualisation stuff
import matplotlib.pyplot as plt

# Histogram or plot various visulaizatioon 
data.HouseAge.hist()
plt.title('House Age')
plt.show()


# now for full data set histogram as example we can 
# show it nicely like this

data.hist(figsize=(12,8))
plt.tight_layout()
```
