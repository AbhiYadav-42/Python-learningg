```python
# doing some greater conditional stuff

def myfunction(x):
    if x % 3 == 0:
        return x ** 2
    else:
        return x//2
```

```python
def function2(x):
    if x.endswith('r'):
        return 'without job'
    else: 
        return x
```

```python
data1.age = data1.age.apply(myfunction)
```

```python
data1.job = data1.job.apply(function2)
```

```python
data1
```

```python
data1['summary'] =  data1.apply(lambda row: f'Age: {row["age"]}, job: {row["job"]}', axis =1)
```

```python
data1.drop('summary', axis=1)
```

```python
# Dropiing a row
data1.drop(['age', 'job'] ,  axis=1)
```
